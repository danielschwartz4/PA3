```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline
```

```python
def read_file(algo_name):
    file_name = "../experiments/" + algo_name + ".csv"
    return pd.read_csv(file_name, sep="\t")
```

```python
kk = read_file("kk")
repeated_random = read_file("repeated_random")
hill_climb = read_file("hill_climbing")
simulated_annealing = read_file("simulated_annealing")
par_repeated_random = read_file("par_repeated_random")
par_hill_climb = read_file("par_hill_climbing")
par_simulated_annealing = read_file("par_simulated_annealing")
```

```python
residues = pd.concat([kk.residue, repeated_random.residue, hill_climb.residue,simulated_annealing.residue,
                     par_repeated_random.residue, par_hill_climb.residue, par_simulated_annealing.residue],axis=1)
residues.columns = ["kk", "repeated_random", "hill_climbing","simulated_annealing",
                   "partition_repeated_random", "partition_hill_climbing","par_simulated_annealing"]
```

```python
residues.boxplot(figsize=(10,6))
plt.title("residue")
plt.xticks(rotation = 90)
plt.ylim((-1,6.052179e15))
```

```python
time = pd.concat([kk.time_spent, repeated_random.time_spent, hill_climb.time_spent,simulated_annealing.time_spent,
                     par_repeated_random.time_spent, par_hill_climb.time_spent, par_simulated_annealing.time_spent],axis=1)
time.columns = ["kk", "repeated_random", "hill_climbing","simulated_annealing",
                   "partition_repeated_random", "partition_hill_climbing","par_simulated_annealing"]
```

```python
time.boxplot(figsize=(10,6))
plt.title("time spent")
plt.xticks(rotation = 90)
plt.ylabel("ms")
```

```python
time.mean().plot(kind="bar", figsize=(10,6))
plt.title("time spent")
plt.ylabel("ms")
```
